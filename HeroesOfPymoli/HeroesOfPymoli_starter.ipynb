### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
total_player = len(purchase_data["SN"].unique())

print(f"Total number of players: {total_player}")

purchase_data.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python

total_unique_items = len(purchase_data["Item Name"].unique())

avg_price =  purchase_data["Price"].mean()

avg_price = round(avg_price, 2)

num_purchase =len(purchase_data["Purchase ID"])

total_revenue =  purchase_data["Price"].sum()

total_revenue = round(total_revenue, 2)


summary_table = pd.DataFrame({"Number of Unique Items": total_unique_items,
                              "Average Price ($)": [avg_price],
                              "Number of Purchases":[num_purchase],
                              "Total Revenue ($)": [total_revenue]})
summary_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python

M = len(purchase_data.loc[purchase_data["Gender"]=="Male", ["Gender"]])

F = len(purchase_data.loc[purchase_data["Gender"]=="Female", ["Gender"]])

total = len(purchase_data["Gender"])

MP = round((M/total)*100, 1)

FP = round((F/total)*100, 1)

NN = total - (M+F)

NNP = round((NN/total)*100, 1)

#summary_table1 = pd.DataFrame({"Total Count": total, "M": [M], "F": [N], "Other/Non-Disclosed": [NN], "Percentage of M": [MP], "Percentage of F": [NP], "Percentage of others": [NNP]})

df = pd.DataFrame({'':["Male", "Female", "Other"], 'Total count':[M,F, NN], 'Percentage of Players (%)':[MP,FP, NNP]})

df.set_index("")
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

pcm = len(purchase_data.loc[purchase_data["Gender"]=="Male", ["Purchase ID"]])

pcf = len(purchase_data.loc[purchase_data["Gender"]=="Female", ["Purchase ID"]])

pco = num_purchase - (pcm+pcf)


summ = purchase_data.loc[purchase_data["Gender"]=="Male", ["Price"]].sum()

summ = summ.iloc[0]

asumm = round(summ/pcm, 2)

sumf = purchase_data.loc[purchase_data["Gender"]=="Female", ["Price"]].sum()

sumf = sumf.iloc[0]

asumf = round(sumf/pcf, 2)

sumt = purchase_data["Price"].sum()

sumnn = sumt - (summ + sumf)

asumnn = round(sumnn/pco, 2)


df = pd.DataFrame({'':["Male", "Female", "Other"], 'Purchase count':[pcm,pcf, pco], 'Total Purchase Value ($)':[summ,sumf, sumnn], 'Average Purchase Price ($)':[asumm, asumf, asumnn]})

df.set_index("")

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 10, 14, 19, 24, 29, 34, 39, 100]

group_names = ["<10", "10-14","15-19", "20-24", "25-30", "30-34", "34-39", "40+"]


purchase_data["View Group"] = pd.cut(purchase_data["Age"], bins, labels=group_names)

purchase_data.head()

p_group = purchase_data.groupby("View Group")

info1_pd = p_group[["SN"]].count() 

info2_pd = round((p_group[["SN"]].count()/total_player)*100, 1)


merge_table = pd.merge(info1_pd, info2_pd, on="View Group")
merge_table

fin_table =merge_table.rename(columns = {"SN_x":"Total Count", "SN_y":"Percentage of Players (%)"})

fin_table
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
info3_pd = p_group[["Purchase ID"]].count() 

info4_pd = p_group[["Price"]].sum() 

info5_pd = round(p_group[["Price"]].mean(), 2)

merge_table1 = pd.merge(info3_pd, info4_pd, on="View Group")
merge_table1

merge_table2 = pd.merge(merge_table1, info5_pd, on="View Group")

fin2_table = merge_table2.rename(columns = {"Purchase ID":"Purchase Count", "Price_x":"Total Purchase Value","Price_y":"Average Purchase Value"})

fin2_table
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order

```python
p_d = pd.read_csv(file_to_load)

g_group = p_d.groupby("SN")

info6_pd = g_group[["Purchase ID"]].count()

info7_pd = g_group[["Price"]].sum()

info8_pd = round(g_group[["Price"]].mean(), 2)

merge_table3 = pd.merge(info6_pd, info7_pd, on="SN")
merge_table3

merge_table4 = pd.merge(merge_table3, info8_pd, on="SN")

fin3_table = merge_table4.rename(columns = {"Purchase ID":"Purchase Count", "Price_x":"Total Purchase Value","Price_y":"Average Purchase Value"})

fin3_table.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
p_d = pd.read_csv(file_to_load)

g_group = p_d.groupby("Item ID")

info6_pd = g_group[["Purchase ID"]].count()

info7_pd = g_group[["Price"]].sum()

info8_pd = round(g_group[["Price"]].mean(), 2)

merge_table3 = pd.merge(info6_pd, info7_pd, on="Item ID")
merge_table3


merge_table4 = pd.merge(merge_table3, info8_pd, on="Item ID")

fin3_table = merge_table4.rename(columns = {"Purchase ID":"Purchase Count", "Price_x":"Total Purchase Value","Price_y":"Average Purchase Value"})

fin3_table.head()
```
